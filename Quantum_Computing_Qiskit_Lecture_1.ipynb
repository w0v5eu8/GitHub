```python
# GHZ       .
def ghz_circuit(n: int):
    assert isinstance(n, int) and n > 0
    qc = QuantumCircuit(n)
    qc.h(0)
    if n>1:
        qc.cx(0, range(1,n))
    qc.measure_all()
    return qc
```

## Deutsch-Jozsa 

Deutsch-Jozsa      .

     .

n     0  1    $f:\{0,1\}^n \rightarrow \{0,1\}$       .

1)     , (constant)

2)     0,   1  . (balanced)

    $f$          .

```python
# Deutsch-Jozsa   .

def dj_circuit(n, f_type):
    num_qubits = n+1
    qc = QuantumCircuit(num_qubits)

    for i_qubit in range(num_qubits-1):
        qc.h(i_qubit)

    qc.x(num_qubits-1)
    qc.h(num_qubits-1)

    qc.barrier()

    if f_type == 'constant':
        pass
    elif f_type == 'balanced':
        for i_qubit in range(n):
            qc.cx(i_qubit, n)
    else:
        raise RuntimeError('Unknown function type')

    qc.barrier()

    for i_qubit in range(num_qubits):
        qc.h(i_qubit)

    qc.measure_all()

    return qc
```

```python
qc = dj_circuit(4, 'constant')
qc.draw('mpl')
```

```python
qc = dj_circuit(4, 'balanced')
qc.draw('mpl')
```
