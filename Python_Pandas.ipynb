```
import pandas as pd
agg_test_df = pd.DataFrame({'key' : ['A','A','A','B','B','B','C','C'],
                   'data' : [1,4,5,2,7,8,3,6]})

agg_test_df_result = agg_test_df.groupby('key').agg(lambda x:x.sum())
agg_test_df_result
```

```
import pandas as pd
agg_test_df = pd.DataFrame({'key' : ['A','A','A','B','B','B','C','C'],
                   'data' : [1,1,2,3,4,4,7,7]})

agg_test_df_result = agg_test_df.groupby('key').agg(lambda x:x.value_counts().index[0])
agg_test_df_result
```
