```python
# Deutsch's Algorithm

def constant():
  circuit = QuantumCircuit(2)
  circuit.cx(1,0)
  return circuit

def balanced():
  circuit = QuantumCircuit(2)
  circuit.cx(0,1)
  return circuit
```

```python
circuit = QuantumCircuit(2,1)
circuit.h(0)
circuit.x(1)
circuit.h(1)
circuit.barrier()
circuit = circuit.compose(constant())
circuit.barrier()
circuit.h(0)
circuit.measure([0],[0])
circuit.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
results = execute(circuit, backend).result()
counts = results.get_counts()

print(counts)
plot_histogram(counts)
```

```python
circuit = QuantumCircuit(2,1)
circuit.h(0)
circuit.x(1)
circuit.h(1)
circuit.barrier()
circuit = circuit.compose(balanced())
circuit.barrier()
circuit.h(0)
circuit.measure([0],[0])
circuit.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
results = execute(circuit, backend).result()
counts = results.get_counts()

print(counts)
plot_histogram(counts)
```

```python
# Grover's Algorithm

def init(n):
  circuit = QuantumCircuit(n+1)
  for i in range(n):
    circuit.h(i)
  circuit.x(n)
  circuit.h(n)
  return circuit

def oracle(n,x):
  circuit = QuantumCircuit(n+1)
  for i in range(n):
    if x & (1 << i) == 0:
      circuit.x(i)
    if n == 2:
      circuit.ccx(0,1,2)
    for i in range(n):
      if x & (1 << i) == 0:
        circuit.x(i)
    return circuit

def diffusion(n, x):
  circuit = QuantumCircuit(n+1)
  for i in range(n):
    circuit.h(i)
    circuit.x(i)
  circuit.h(n)
  if n ==2:
    circuit.h(1)
    circuit.cx(0,1)
    circuit.h(1)
  for i in range(n):
    circuit.x(i)
    circuit.h(i)
  return circuit

n=2
x=3
circuit=QuantumCircuit(3,2)
circuit = circuit.compose(init(n))
circuit.barrier()
circuit = circuit.compose(oracle(n,x))
circuit.barrier()
circuit = circuit.compose(diffusion(n,x))
circuit.barrier()
bits=[i for i in range(n)]
circuit.measure(bits, bits)
circuit.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
results = execute(circuit, backend).result()
counts = results.get_counts()

print(counts)
plot_histogram(counts)
```
